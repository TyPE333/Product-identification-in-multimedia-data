### Exploring styles.csv

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
styles = pd.read_csv("../data/raw/styles.csv",on_bad_lines='skip')
```

```python
styles.head(10)
```

```python
styles.isna().sum()
```

```python
styles["image_path"] = styles["id"].astype(str).apply(lambda x: f"../data/raw/images/{x}.jpg")
styles["product_url"] = styles["productDisplayName"].astype(str).apply(lambda x: f"https://www.example.com/product/{x}.")
```

```python
styles['baseColour'].fillna('Unknown', inplace=True)
styles['season'].fillna('Unknown', inplace=True)
styles['usage'].fillna('Unknown', inplace=True)
styles['year'].fillna(styles['year'].median(), inplace=True)
styles.dropna(subset=['productDisplayName'], inplace=True)
```

#### Additional Preprocessing

```python
styles
```

```python
#Strip whitespace, lower case, remove special characters
styles["productDisplayName"] = styles["productDisplayName"].str.strip().str.lower().str.replace(r"[^a-zA-Z0-9/s]", "", regex=True)

#Convert to categorical
for col in ['gender', 'masterCategory', 'subCategory', 'articleType', 'baseColour', 'season', 'usage']:
    styles[col] = styles[col].astype('category')

#year to int
styles["year"] = styles["year"].astype(int)

#Verify that all image files listed in styles.csv exist in the images/ folder
import os
img_path = 'C:/Users/rohan/OneDrive/Documents/Downloads/Resume projects april 2025/Product-identification-in-multimedia-data/data/raw/images'
styles['imageExists'] = styles['id'].apply(lambda x: os.path.exists(os.path.join(img_path, f"{x}.jpg")))
#styles = styles[styles['imageExists']]
```
